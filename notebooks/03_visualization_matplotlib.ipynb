```python
# Import necessary library
import matplotlib.pyplot as plt
```

Introduction
Matplotlib is a powerful plotting library in Python used for creating static, interactive, and animated visualizations. It helps us understand patterns, trends, and distributions in data.

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Line plot
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Categories
subjects = ['Math', 'Science', 'History', 'English']
scores = [88, 92, 79, 85]

plt.bar(subjects, scores, color='skyblue')
plt.title("Bar Plot - Student Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```

```python
# Random data
import numpy as np
data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title("Histogram - Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 15, 10, 25, 30]

plt.scatter(x, y, color='green')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
labels = ['Python', 'R', 'SQL', 'Others']
sizes = [45, 25, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Programming Language Usage")
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='red', linestyle='--', linewidth=2, label='sin(x)')
plt.title("Customized Line Plot")
plt.xlabel("x values")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Sin(x)")

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x), color='orange')
plt.title("Cos(x)")

plt.tight_layout()
plt.show()
```
