```python
# Import necessary library
import matplotlib.pyplot as plt
```

Introduction
Matplotlib is a powerful plotting library in Python used for creating static, interactive, and animated visualizations. It helps us understand patterns, trends, and distributions in data.

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Line plot
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Categories
subjects = ['Math', 'Science', 'History', 'English']
scores = [88, 92, 79, 85]

plt.bar(subjects, scores, color='skyblue')
plt.title("Bar Plot - Student Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```

```python
# Random data
import numpy as np
data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title("Histogram - Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
